```python
import pandas as pd

url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"

df = pd.read_csv(url, sep=',')

df.head()
```

```python
df = df.drop('url', axis=1)
display(df.head())

#Drop url karena tidak ada fungsinya
```

```python
# Menghitung null di setiap kolom
null_count = df.isnull().sum()

# Menghitung persentase null di kolom
null_percentage = (null_count / len(df)) * 100

# Buat dataframe untuk menampilkan
null_df = pd.DataFrame({'Null Count': null_count, 'Null Percentage %': null_percentage})

# Memfilter kolom yang ga ada null dan di sort null count
null_info = null_df[null_df['Null Count'] > 0].sort_values(by='Null Count', ascending=False)

# Menampilkan
print(null_info)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot
plt.figure(figsize=(12,6))
sns.barplot(x='Null Count', y=null_info.index, data=null_info, orient='h', palette='rainbow')
plt.title('Number of Null per Column')
plt.xlabel('Null Count')
plt.ylabel('Column Name')

# Tambahkan null count
for index, row in null_info.iterrows():
    plt.text(row['Null Count'] + 1, index, str(row['Null Count']), ha='left', va='center')
plt.show()
```

```python
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 30)
```

```python
# Handle missing values
if df['property_type'].isnull().sum() > 0:
    df['property_type'].fillna(df['property_type'].mode()[0], inplace=True)

df['ads_id'].fillna("unknown", inplace=True)
df['bedrooms'].fillna(df['bedrooms'].median(), inplace=True)
df['bathrooms'].fillna(df['bathrooms'].median(), inplace=True)
df['land_size_m2'].fillna(df['land_size_m2'].median(), inplace=True)
df['building_size_m2'].fillna(df['building_size_m2'].median(), inplace=True)
df['certificate'].fillna(df['certificate'].mode()[0], inplace=True)
df['floors'].fillna(df['floors'].mode()[0], inplace=True)

median_age = df['building_age'].median()

if pd.isna(median_age):
    df['building_age'] = 10  # fallback default jika semua kosong
else:
    df['building_age'].fillna(median_age, inplace=True)

# Hitung year_built dari building_age jika kosong
if 'year_built' in df.columns:
    df['year_built'] = df['year_built'].fillna(2025 - df['building_age'])


df['property_condition'].fillna(df['property_condition'].mode()[0], inplace=True)
df['building_orientation'].fillna("unknown", inplace=True)
df['furnishing'].fillna("unspecified", inplace=True)

print("Missing values after handling:")
print(df.isnull().sum())
```

```python
# Buang outlier yang terlalu jauh
numeric_cols = df.select_dtypes(include=['float64','int64']).columns
outlier_removed = df.copy()

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    before = outlier_removed.shape[0]
    outlier_removed = outlier_removed[(outlier_removed[col] >= lower_bound) & (outlier_removed[col] <= upper_bound)]
    after = outlier_removed.shape[0]

    print(f"{col:<20}: Removed {before - after} rows (kept {after} rows)")

df = outlier_removed
print("Data yang tidak masuk akal sudah dihapus, sisa yang masuk akal saja")
```

```python
import numpy as np

# 1. Pastikan building_age terisi
if df['building_age'].dropna().shape[0] == 0:
    df['building_age'] = 10
else:
    df['building_age'].fillna(df['building_age'].median(), inplace=True)

# 2. Pastikan land_size_m2 > 0 (hindari pembagian dengan nol)
df = df[df['land_size_m2'] > 0]

# 3. Price category bins
price_bins = [0, 500_000_000, 1_500_000_000, 3_000_000_000, df['price_in_rp'].max()+1]
price_labels = ['Low','Medium','High','Luxury']
df['price_category'] = pd.cut(df['price_in_rp'], bins=price_bins, labels=price_labels, include_lowest=True)

# 4. Building age bins
max_building_age = df['building_age'].max()
if np.isnan(max_building_age):
    max_building_age = 30

age_bins = [0, 5, 10, 20, max_building_age + 1]
age_bins = sorted(list(set(bin for bin in age_bins if bin >= 0)))  # memastikan berurutan & unik
age_labels = [f'{age_bins[i]}-{age_bins[i+1]-1} yrs' for i in range(len(age_bins)-1)]
df['age_group'] = pd.cut(df['building_age'], bins=age_bins, labels=age_labels, include_lowest=True)

# 5. Price per m² (hindari NaN)
df['price_per_m2'] = df['price_in_rp'] / df['land_size_m2']
df['price_per_m2'].fillna(df['price_per_m2'].median(), inplace=True)

print("=== Feature Engineering Done ===")
print("Distribusi age_group:")
print(df['age_group'].value_counts(dropna=False))
display(df[['price_in_rp','price_category','building_age','age_group','price_per_m2']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Pastikan price_category & age_group terbentuk
if 'price_category' not in df.columns or df['price_category'].isna().all():
    price_bins = [0, 500_000_000, 1_500_000_000, 3_000_000_000, df['price_in_rp'].max()+1]
    price_labels = ['Low','Medium','High','Luxury']
    df['price_category'] = pd.cut(df['price_in_rp'], bins=price_bins, labels=price_labels, include_lowest=True)

if 'age_group' not in df.columns or df['age_group'].isna().all():
    max_building_age = df['building_age'].max()
    if np.isnan(max_building_age):
        max_building_age = 30
    age_bins = [0, 5, 10, 20, max_building_age + 1]
    age_bins = sorted(list(set(bin for bin in age_bins if bin >= 0)))
    age_labels = [f'{age_bins[i]}-{age_bins[i+1]-1} yrs' for i in range(len(age_bins)-1)]
    df['age_group'] = pd.cut(df['building_age'], bins=age_bins, labels=age_labels, include_lowest=True)

# 1. Distribusi Harga
plt.figure(figsize=(10,5))
sns.histplot(df['price_in_rp'], bins=60, kde=True)
plt.title("Distribusi Harga Rumah")
plt.xlabel("Harga (Rp)")
plt.ylabel("Jumlah")
plt.show()
print("Insight: Distribusi harga miring ke kanan, mayoritas rumah di bawah Rp 1.5 miliar.")

# 2. Harga per Kota
plt.figure(figsize=(12,5))
order = df.groupby('city')['price_in_rp'].median().sort_values(ascending=False).index
sns.boxplot(x='city', y='price_in_rp', data=df, order=order)
plt.xticks(rotation=45)
plt.title("Harga Rumah per Kota")
plt.show()
print("Insight: Kota dengan median harga tertinggi adalah Jakarta, sedangkan Bekasi dan Depok cenderung lebih murah.")

# 3. Jumlah Rumah per Kategori Harga
plt.figure(figsize=(6,4))
sns.countplot(x='price_category', data=df, order=['Low','Medium','High','Luxury'])
plt.title("Jumlah Rumah per Kategori Harga")
plt.xlabel("Kategori Harga")
plt.ylabel("Jumlah Properti")
plt.show()
print("Insight: Sebagian besar rumah berada di kategori Low dan Medium, menandakan pasar didominasi rumah yang harganya lebih terjangkau.")

# 4. Distribusi Umur Bangunan
plt.figure(figsize=(6,4))
sns.countplot(x='age_group', data=df, order=age_labels)
plt.title("Distribusi Umur Bangunan")
plt.xlabel("Kelompok Umur Bangunan")
plt.ylabel("Jumlah Properti")
plt.show()
print("Insight: Mayoritas rumah relatif baru (<= 10 tahun).")

# 5. Heatmap Korelasi
plt.figure(figsize=(10,6))
selected_numeric = ['price_in_rp','land_size_m2','building_size_m2','bedrooms','bathrooms','price_per_m2']
sns.heatmap(df[selected_numeric].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap Korelasi Kolom Numerik")
plt.show()
print("Insight: Harga memiliki korelasi kuat dengan luas tanah dan luas bangunan, sedangkan korelasi dengan kamar tidur/bathroom lebih lemah.")

# 6. Scatterplot Harga vs Luas Tanah
plt.figure(figsize=(8,4))
sample_df = df.sample(min(1000, len(df)), random_state=1)  # aman untuk dataset kecil
sns.scatterplot(x='land_size_m2', y='price_in_rp', data=sample_df, alpha=0.6)
plt.title("Harga vs Luas Tanah")
plt.xlabel("Luas Tanah (m²)")
plt.ylabel("Harga (Rp)")
plt.show()
print("Insight: Semakin luas tanah, harga semakin tinggi, tetapi ada properti kecil yang tetap mahal (kemungkinan lokasi strategis/premium).")
```
